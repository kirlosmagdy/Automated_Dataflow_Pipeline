# Prepare Restaurants Data
### 1- add city column for each file 
### 2- merge both csv files in 1 csv file called restaurants
### 3- add id column after merging the files


```python
import pandas as pd
```

```python
# Load raw files
files = {
    "alexandria"    : r"D:\Mashro3 El T5arog\pipeline\scraping_scripts\Data_files\alex_2026-05-02_talabat.csv",
    "cairo"         : r"D:\Mashro3 El T5arog\pipeline\scraping_scripts\Data_files\cairo_2026-05-02_talabat.csv",
}
 
dfs = {city: pd.read_csv(path) for city, path in files.items()}
```

```python
# 1- add city column to the data frame
for city, df in dfs.items():
    df["City"] = city
```

```python
# 2- merge all data frames into one file 
merged_df = pd.concat(dfs.values(), ignore_index=True)
 
merged_df.to_csv("talabat_merged.csv", index=False)
```

```python
# 3- add id column to the merged data frame
merged_df.insert(0, "id", range(1, len(merged_df) + 1))   # integer IDs
```

```python
# Move City to second column (right after Hotel_ID)
cols = list(merged_df.columns)
cols.insert(1, cols.pop(cols.index("City")))
merged_df = merged_df[cols]
```

```python
# save the final merged data frame to a new CSV file in database folder
merged_df.to_csv(r"D:\\Mashro3 El T5arog\\pipeline\\database\\Data_files\\restaurants.csv", index=False)
```
